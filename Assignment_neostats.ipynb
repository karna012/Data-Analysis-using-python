```python
outliers = {}
for column in df.columns:
    if df[column].dtype != 'object':
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_threshold = Q1 - 1.5 * IQR
        upper_threshold = Q3 + 1.5 * IQR
        outliers[column] = df[(df[column] < lower_threshold) | (df[column] > upper_threshold)][column].tolist()
```

```python
print("Outliers:\n", outliers)
```

```python
from scipy.stats import chi2_contingency
```

```python
for column in str_col:
    contingency_table = pd.crosstab(df[column], df['class']).values
    chi2_stat, p_value, dof, expected_freq = chi2_contingency(contingency_table)
    print(f"Contingency table for {column}:")
    print(pd.crosstab(df[column], df['class']))
    print(f"Chi-squared test results: chi2_stat={chi2_stat}, p_value={p_value}, dof={dof}, expected_freq={expected_freq}")
```

```python
X_train, X_test, y_train, y_test = train_test_split(final_df2.drop('class', axis=1), final_df2['class'], test_size=0.3, random_state=42)


```

```python
print("Accuracy:", accuracy_score(y_test, lr_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, lr_pred))
print("\nClassification Report:\n", classification_report(y_test, lr_pred))
```
