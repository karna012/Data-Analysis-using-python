```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
df = pd.read_excel('superstore_sales.xlsx')
```

```python
df['product_name'].nunique()
```

```python
top_10_b_s_p = df.groupby('product_name').agg({'quantity':'sum'}).sort_values('quantity', ascending = False)[:10]
top_10_b_s_p
```

```python
sns.barplot(x='product_name', y='quantity', data=top_10_b_s_p.reset_index())
plt.xticks(rotation = 'vertical')
```

```python
productSales = pd.DataFrame(df.groupby('product_name').sum()['sales'])
productSales.sort_values(by=['sales'], inplace=True, ascending=False)
productSales.head(10)
```

```python
most_profit = pd.DataFrame(df.groupby('product_name').sum()['profit'])
most_profit.sort_values(by=['profit'], inplace = True, ascending = False)
most_profit.head(10)
```

```python
sns.barplot(x='product_name', y='profit', data = most_profit.head(10).reset_index())
plt.xticks(rotation='vertical')
```

```python
sns.barplot(x='product_name', y='sales', data = productSales.head(10).reset_index())
plt.xticks(rotation = 'vertical')
```
