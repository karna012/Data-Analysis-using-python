```python
def convert_date(date_str):
    """Converts date string to datetime object"""
    try:
        date = datetime.strptime(date_str, '%m/%d/%Y %H:%M')
    except ValueError:
        date = datetime.strptime(date_str, '%d-%m-%Y %H:%M')
    return date

```

```python
df['Date'] = df['Date/Time'].apply(convert_date)
```

```python
variance = df['Rel Hum_%'].var()
Mean = df['Rel Hum_%'].mean()
Std_Dev = df['Rel Hum_%'].std()
```

```python
Std_Dev = df['Rel Hum_%'].std()
Mean + Std_Dev
```

```python
plt.plot(df['Rel Hum_%'], 'm')
plt.axhline(Mean, color='red', label='Mean')
plt.axhline(Mean + Std_Dev, color='green', linestyle='dashed', label='Std Dev')
plt.axhline(Mean - Std_Dev, color='green', linestyle='dashed')
plt.axhline(Mean + 2*Std_Dev, color='orange', linestyle='dashed', label='2 Std Dev')
plt.axhline(Mean - 2*Std_Dev, color='orange', linestyle='dashed')
plt.axhline(Mean + 3*Std_Dev, color='purple', linestyle='dashed', label='3 Std Dev')
plt.axhline(Mean - 3*Std_Dev, color='purple', linestyle='dashed')
plt.title('Rel Hum_% Scatter Plot')
plt.xlabel('Index')
plt.ylabel('Rel Hum_%')
plt.legend()
plt.show()
```
